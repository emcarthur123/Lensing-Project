```python
import matplotlib.pyplot as plt
import numpy as np

# Generating some data for demonstration
x_values = np.linspace(0, 2, 100)
y_values = np.linspace(0, 2, 100)

# Define darker shades of colors
dark_red = '#8B0000'  # Dark Red
dark_blue = '#00008B'  # Dark Blue
dark_green = '#006400'  # Dark Green

fig, axs = plt.subplots(2, 3, figsize=(12, 8), sharey=True)

# Plotting high lensing data
scatter1 = axs[0, 0].scatter(elg_redshift_high_machine, pred_red_high_machine, color=dark_red, label='High SNR')
axs[0, 0].plot(x_values, x_values, linestyle='--', color='black')
axs[0, 0].fill_between(x_values, x_values - 0.1, x_values + 0.1, color='gray', alpha=0.3)  # Add shaded region
axs[0, 0].set_xlabel('ELG Z$_{True}$', fontsize=15)
axs[0, 0].set_ylabel('CNN ELG $Z_{Pred}$', fontsize=15)
axs[0, 0].set_title('High SNR$_{ELG}$', fontsize=15)


# Plotting mid lensing data
scatter2 = axs[0, 1].scatter(elg_redshift_mid_machine, pred_red_mid_machine, color=dark_blue, label='mid SNR')
axs[0, 1].plot(x_values, x_values, linestyle='--', color='black')
axs[0, 1].fill_between(x_values, x_values - 0.1, x_values + 0.1, color='gray', alpha=0.3)  # Add shaded region
axs[0, 1].set_xlabel('ELG Z$_{True}$', fontsize=15)
axs[0, 1].set_title('Mid SNR$_{ELG}$', fontsize=15)

# Plotting low lensing data
scatter3 = axs[0, 2].scatter(elg_redshift_low_machine, pred_red_low_machine, color=dark_green, label='low SNR')
axs[0, 2].plot(x_values, x_values, linestyle='--', color='black')
axs[0, 2].fill_between(x_values, x_values - 0.1, x_values + 0.1, color='gray', alpha=0.3)  # Add shaded region
axs[0, 2].set_xlabel('ELG Z$_{True}$', fontsize=15)
axs[0, 2].set_title('Low SNR$_{ELG}$', fontsize=15)

# Plotting high lensing data
scatter4 = axs[1, 0].scatter(elg_redshift_high_red, pred_red_high_red, color=dark_red, label='High SNR')
axs[1, 0].plot(x_values, x_values, linestyle='--', color='black')
axs[1, 0].fill_between(x_values, x_values - 0.1, x_values + 0.1, color='gray', alpha=0.3)  # Add shaded region
axs[1, 0].set_xlabel('ELG Z$_{True}$', fontsize=15)
axs[1, 0].set_ylabel('Redrock ELG $Z_{Pred}$', fontsize=15)

# Plotting mid lensing data
scatter5 = axs[1, 1].scatter(elg_redshift_mid_red, pred_red_mid_red, color=dark_blue, label='mid SNR')
axs[1, 1].plot(x_values, x_values, linestyle='--', color='black')
axs[1, 1].fill_between(x_values, x_values - 0.1, x_values + 0.1, color='gray', alpha=0.3)  # Add shaded region
axs[1, 1].set_xlabel('ELG Z$_{True}$', fontsize=15)

# Plotting low lensing data
scatter6 = axs[1, 2].scatter(elg_redshift_low_red, pred_red_low_red, color=dark_green, label='low SNR')
axs[1, 2].plot(x_values, x_values, linestyle='--', color='black')
axs[1, 2].fill_between(x_values, x_values - 0.1, x_values + 0.1, color='gray', alpha=0.3)  # Add shaded region
axs[1, 2].set_xlabel('ELG Z$_{True}$', fontsize=15)

# Set label for color bars
#cbar1.set_label('QSO Redshift')
#cbar2.set_label('QSO Redshift')

# Adjust layout to prevent overlap and remove spacing between plots
fig.subplots_adjust(wspace=0, hspace=0)

# Display the plot
plt.show()
```
